3. Напишите функцию, которая принимает матрицу и производит нормировку её элементов (т.е. $A'_{ij} = A_{ij} / \sum_{i} \sum_{j} A_{ij}$)

```python
def normalize_matrix(matrix):
    total_sum = np.sum(matrix)
    normalized_matrix = matrix / total_sum
    return normalized_matrix

test_matrix = np.arange(0, 17).reshape(4, 4)
print(test_matrix)
print(normalize_matrix(test_matrix))
```

4. Напишите функцию, которая принимает 3 трехмерных вектора $x, y, z$ и вычисляет объем пирамиды, образованной этими векторами (используйте смешанное произведение векторов)

```python
def pyramid_volume(v1, v2, v3):
    mixed_product = np.dot(v1, np.cross(v2, v3))
    volume = abs(mixed_product) / 6
    return volume

v1 = np.array([1, 0, 0])
v2 = np.array([0, 1, 0])
v3 = np.array([0, 0, 1])

print(pyramid_volume(v1, v2, v3))
```

5. Напишите функцию, которая принимает число N и строит матрицу NxN, с нулями внутри и единицами на границе.

```python
def border_matrix(N):
    matrix = np.ones((N, N), dtype=int)
    if N > 2:
        matrix[1:N-1, 1:N-1] = 0
    return matrix

print(border_matrix(10))
```

6. Создайте матрицу шахматную доску 8х8 используя функцию `tile`

```python
def chessboard():
    pattern = np.array([[0, 1],
                        [1, 0]])
    board = np.tile(pattern, (4, 4))
    
    return board
new_board = chessboard()
print(new_board)
```

7. Как найти наиболее близкое значение (к заданному скаляру) в векторе?

```python
def find_closest(arr, value):
    diff = np.abs(arr - value)
    i = np.argmin(diff)
    return arr[i]

test_arr = np.arange(0, 16)
val = 14.44444
print(find_closest(test_arr, val))
```

10. Даны 2 массива Х и Y, постройте матрицу Коши $C$ ($C_{ij} = \frac{1}{x_i - y_j}$)

```python
def c_matrix(X, Y):
    C = 1.0 / (X[:, np.newaxis] - Y[np.newaxis, :])
    return C

X_matrix = np.ones((3, 3))
Y_matrix = np.arange(50, 59).reshape(3, 3)
print(c_matrix(Y_matrix, X_matrix))
```

### 3. Визуализация

Используя библиотеки matplotlib/sns/plotly/pandas и набор данных представленный ранее визуализируйте

1. Сделайте круговую диаграмму, на которой были бы отражено число заболеваний COVID-19 по континентам

```python
import pandas as pd
import matplotlib.pyplot as plt

cases_by_continent = merged_data.groupby('continent')['total_cases'].sum()
plt.figure(figsize=(8, 8))
plt.pie(cases_by_continent, labels=cases_by_continent.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Число заболеваний COVID-19 по континентам')
plt.axis('equal')
plt.show()
```

2. Постройте диаграмму, на которой в порядке убывания были бы отмечены число тестов на COVID-19 в России, Китае, США, Индии и Нигерии

```python
merged_data['total_tests'].fillna(0, inplace=True)
countries = ['Russia', 'China', 'United States', 'India', 'Nigeria']
filtered_data = merged_data[merged_data['location'].isin(countries)]
sorted_data = filtered_data.sort_values(by='total_tests', ascending=False)
plt.figure(figsize=(10, 6))
plt.bar(sorted_data['location'], sorted_data['total_tests'], color='skyblue')
plt.title('Число тестов на COVID-19 по странам')
plt.xlabel('Страны')
plt.ylabel('Количество тестов')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
